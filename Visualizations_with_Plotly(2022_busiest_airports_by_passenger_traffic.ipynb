```python
#Insert latitude and longitude into the dataframe
df.insert(loc=1, column='long', value=longitudes)
df.insert(loc=1, column='lat', value=latitudes)
```

```python
df
```

```python
# Create a text column for hover information
df['text'] = df['Airport'] + ', ' + df['Location'] + ' (' + df['Country'] + ')<br>' + \
             'Passengers: ' + df['Total passengers'].astype(str) + '<br>' + \
             'Rank: ' + df['Rank'].astype(str)
```

```python
# Create the scattergeo plot
fig = go.Figure(data=go.Scattergeo(
    lon=df['long'],
    lat=df['lat'],
    text=df['text'],
    mode='markers',
    marker=dict(
        size=10,  # Adjust the size of the
        symbol='circle',  # Use a circle symbol for the markers
        color=df['Rank'],  # Color by rank
        colorscale='Viridis',  # Change color scale
        colorbar=dict(title='Rank'),
        line=dict(width=1, color='black')  # Outline color of markers
    )
))
```

```python
# Customize the layout
fig.update_layout(
    title='Top 50 Busiest Airports Worldwide (2022)',
    geo=dict(
        scope='world',  # Show the entire world map
        projection=go.layout.geo.Projection(type='natural earth'),  # Use a natural earth projection
        showland=True,  # Show land
        landcolor='rgb(243, 243, 243)',  # Land color
        subunitwidth=1,
        countrywidth=1,
        subunitcolor="rgb(217, 217, 217)",
        countrycolor="rgb(217, 217, 217)"
    )
)

# Show the plot
fig.show()
```
